project name: Image to Pencil Sketch with Python

importing important packages

```python
import cv2
from matplotlib import pyplot as plt
```

```python
sakti = cv2.imread("/content/drive/MyDrive/entrndata/saktiman.png")
color = cv2.cvtColor(sakti, cv2.COLOR_BGR2RGB)
plt.figure(figsize = (10,10))
plt.axis("off")
plt.imshow(color)
plt.title('Shaktimaan  indian superhero')
plt.show
```

greyscalling orginal image

```python
grey = cv2.cvtColor(sakti, cv2.COLOR_BGR2GRAY)
plt.figure(figsize = (10,10))
plt.axis('off')
plt.title('image converted into GreyScale')
plt.imshow(grey,cmap='gray')
plt.show()
```

inverting greyscale image

```python
inverted_img = 255 - grey
plt.figure(figsize = (10,10))
plt.axis('off')
plt.title('Inverted grey scale Image')
plt.imshow(inverted_img,cmap='gray')
plt.show()
```

blurring the image

```python
blur = cv2.GaussianBlur(inverted_img, (21,21),0)
plt.figure(figsize = (10,10))
plt.axis('off')
plt.title('Blurred Image')
plt.imshow(blur,cmap='gray')
plt.show()
```

converting blurred image into pencil sketch

```python
invert = 255 - blur
pencil = cv2.divide(grey,invert,scale=256.0)
plt.figure(figsize = (10,10))
plt.axis('off')
plt.title('Pencil Image')
plt.imshow(pencil,cmap='gray')
plt.show()
```

comparing original image to pencil image

```python
plt.figure(figsize = (10,10))
plt.subplot(1,2,1)
plt.imshow(color)
plt.axis('off')
plt.title('Original Image')
plt.subplot(1,2,2)
plt.imshow(pencil,cmap='gray')
plt.axis('off')
plt.title('Pencil sektech Shaktimaan ')
```
